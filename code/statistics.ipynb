```python
df = pd.read_csv(rf'{export}results.csv', sep=';')

display(df)
```

```python
df2 = df[['MappingMethod', 'NumLayers', 'Target_Relative', 'Result_Numeric_Completed', 'Result_Numeric_Failed', 'Result_Numeric_Terminated']]

display(df2)

df3 = df2.pivot_table(index=['MappingMethod', 'Target_Relative'], aggfunc='sum').reset_index()

df3['Sum'] = df3['Result_Numeric_Completed'] + df3['Result_Numeric_Failed'] + df3['Result_Numeric_Terminated']

# desc = df2.groupby().describe()

display(df3)
```

```python
display(df3['Target_Relative'].astype(float).drop_duplicates())
```

```python
bins = [0, 0.25, 0.51, 0.76, 1.0]
labels = ["<0.25", "<=0.5", "<=0.75", ">0.75"]

df3["Target_Class"] = pd.cut(df3["Target_Relative"], bins=bins, labels=labels, include_lowest=True, right=False)

df_grouped = df3.groupby(["MappingMethod", "Target_Class"])[['Result_Numeric_Completed', 'Result_Numeric_Failed', 'Result_Numeric_Terminated', 'Sum']].sum().reset_index()

display(df_grouped)
```

```python
df_grouped['Result_Numeric_Completed_Relative'] = df_grouped['Result_Numeric_Completed'] / df_grouped['Sum']
df_grouped['Result_Numeric_Failed_Relative'] = df_grouped['Result_Numeric_Failed'] / df_grouped['Sum']
df_grouped['Result_Numeric_Terminated_Relative'] = df_grouped['Result_Numeric_Terminated'] / df_grouped['Sum']

display(df_grouped)
```

```python

# Annahme: df_grouped enthält Spalten: Mapping, Target_Class, Result1, Result2, ...
result_cols = ['Result_Numeric_Completed_Relative', 'Result_Numeric_Failed_Relative', 'Result_Numeric_Terminated_Relative']

# Farbcodes pro Result-Spalte
colors = [color_shades_blue[14], color_shades_purple[10], color_shades_orange[10]]  # Optional: kann erweitert werden

target_classes = df_grouped["Target_Class"].cat.categories

# Liste für Traces
traces = []

# Wir bauen das Diagramm gruppiert nach Mapping + Target_Class
for i, result_col in enumerate(result_cols):
    traces.append(
        go.Bar(
            x=[f"{m}<br>{c}" for m, c in zip(df_grouped["MappingMethod"], df_grouped["Target_Class"])],
            y=df_grouped[result_col],
            name=result_col,
            marker_color=colors[i % len(colors)],
        )
    )

# Plot erzeugen
fig = go.Figure(data=traces)

fig.update_layout(
    barmode="stack",  # Balken nebeneinander
    title="Resultate nach Mapping und Target-Klasse",
    xaxis_title="Mapping + Target-Klasse",
    yaxis_title="Summenwerte",
    xaxis_tickangle=-45
)

fig.show()
```

```python
width=1000
height=500

# Spalten
result_cols = ['Result_Numeric_Completed_Relative', 'Result_Numeric_Failed_Relative', 'Result_Numeric_Terminated_Relative']
result_labels = ['korrekt', 'fehlgeschlagen', 'abgebrochen']
target_classes = df_grouped["Target_Class"].cat.categories
target_labels = ['komprimiert', 'äquidistant', 'gedehnt', '', 'komprimiert ', 'äquidistant ' , 'gedehnt ', ' ', 'komprimiert  ', 'äquidistant  ', 'gedehnt  ' , '   ', 'komprimiert   ', 'äquidistant   ', 'gedehnt   ']
mappings = df_grouped["MappingMethod"].unique()

# Kombinierte X-Achse: Target_Class | Mapping
df_grouped["x"] = df_grouped["Target_Class"].astype(str) + " | " + df_grouped["MappingMethod"]

# X-Achsenreihenfolge erzwingen
x_order = []
idx = 0
for tc in target_classes:
    idx +=1
    for m in mappings:
        label = f"{tc} | {m}"
        if f"{tc} | {m}" in df_grouped["x"].values:
            x_order.append(label)
    x_order.append(' '*idx)

# Gestapelte Traces für jede Result-Spalte
traces = []

for i, result_col in enumerate(result_cols):
    y_values = []
    for x in x_order:
        if x == "":
            y_values.append(0)
        else:
            value = df_grouped.loc[df_grouped["x"] == x, result_col]
            y_values.append(value.values[0] if not value.empty else 0)

    traces.append(go.Bar(
        x=target_labels,
        y=y_values,
        name=result_labels[i],
        marker_color=colors[i] if len(y_values) > 1 else 'rgba(255,255,255,0.0)',
        marker=dict(line=dict(width=1, color='white'))
    ))


layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

# Plot erstellen
fig = go.Figure(layout=layout, data=traces)

fig.update_layout(
    barmode="stack",  # gestapelt
    # title="Gestapelte Balken: Result-Spalten pro Anordnung nach Zieltiefen",
    #xaxis_title="Target_Class | Mapping",
    #yaxis_title="Summierte Werte",
    xaxis_tickangle=-45,
    width=width,
    height=height,
    font_family="Barlow",
    font_color="black",
    title_font_family="Barlow Medium",
    title_font_color="#2c76b8",
    legend_title_font_color="#2c76b8",
    legend_title_font_family='Barlow',
    legend_font_family='Barlow Condensed Thin'
)

fig.update_xaxes(tickwidth=0, tickcolor='rgba(255,255,255,0)', ticklen=10, linecolor='rgba(255,255,255,0)', ticks='outside')
fig.update_yaxes(tickwidth=0, tickcolor='rgba(255,255,255,0)', ticklen=10, linecolor='rgba(255,255,255,0)', ticks='outside')

fig.write_image(rf'{export_img}outcome_mapping-rate_target-classes.svg', width=width, height=height, scale=4)

fig.show()
```
